```python
import yfinance as yf
import pandas as pd
```

## Using the yfinance Library to Extract Stock Data


Using the `Ticker` module we can create an object that will allow us to access functions to extract data. To do this we need to provide the ticker symbol for the stock, here the company is Tesla and the ticker symbol is `TSLA` and GameStop is `GME`


```python
tesla = yf.Ticker("TSLA")
gamestop = yf.Ticker("GME")
```

### Extracting Share Price


A share is the single smallest part of a company's stock  that you can buy, the prices of these shares fluctuate over time. Using the <code>history()</code> method we can get the share price of the stock over a certain period of time. Using the `period` parameter we can set how far back from the present to get data. The options for `period` are 1 day (1d), 5d, 1 month (1mo) , 3mo, 6mo, 1 year (1y), 2y, 5y, 10y, ytd, and max.


```python
tesla_share_price_data = tesla.history(period="30y")
gamestop_share_price_data = gamestop.history(period="30y")
```

The format that the data is returned in is a Pandas DataFrame. With the `Date` as the index the share `Open`, `High`, `Low`, `Close`, `Volume`, and `Stock Splits` are given for each day.


```python
tesla_share_price_data.head()
```

```python
gamestop_share_price_data.head()
```

```python
tesla_revenue = tesla_share_price_data.assign(Revenue = tesla_share_price_data.Close * tesla_share_price_data.Volume)
tesla_revenue.tail()
```

```python
gamestop_revenue = gamestop_share_price_data.assign(Revenue = gamestop_share_price_data.Close * gamestop_share_price_data.Volume)
gamestop_revenue.tail()
```

We can plot the `Open` price against the `Date`:


```python
import matplotlib.pyplot as plt
```

```python
tesla_share_price_data.loc["2023-06-06 00:00:00-04:00","Open"]
```

```python
volumes = tesla_share_price_data['Volume']
close_prices = tesla_share_price_data['Close']
plt.figure(figsize=(10, 6))
plt.plot(volumes, close_prices, color='blue')
plt.xlabel('Volume')
plt.ylabel('Closing Price')
plt.title('Tesla Stock Data')
plt.show()
```
